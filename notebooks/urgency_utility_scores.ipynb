```python
import numpy as np
import pandas as pd
from scipy.stats import gamma
```

```python
donors = pd.read_csv("donors.csv")
patients = pd.read_csv("patients.csv")
print(patients.columns)

patients["BaseUrgency"] = np.random.gamma(2, 5, size=len(patients))
patients["UrgencyScore"] = patients["BaseUrgency"] + patients["Diabetes"] + 0.3 * np.log1p(patients["DialysisYears"]) # 0.3 param can be changed, chosen so that dialysis contributes clearly, but modestly


patients["PostTxSurvival_true"] = 0.88 - 0.0035 * patients["Age"] - 0.010 * patients["UrgencyScore"] - 0.025 * patients["Diabetes"] + np.random.normal(0, 0.03, size=len(patients))
patients["NoTxSurvival_true"] = 0.72 - 0.0035 * patients["Age"] - 0.018 * patients["UrgencyScore"]  - 0.035 * patients["Diabetes"] + np.random.normal(0, 0.05, size=len(patients))

patients["SurvivalBenefit"] = np.maximum(patients["PostTxSurvival_true"] - patients["NoTxSurvival_true"], 0)
print(patients.head())
```
