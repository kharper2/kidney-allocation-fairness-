```python
import numpy as np
import pandas as pd
```

```python
patients = pd.read_csv("patients.csv")
print(patients.columns)

patients["UrgencyScore"] = np.log1p(patients["DialysisYears"] + 1)
patients["UrgencyScore"] = patients["UrgencyScore"] / patients["UrgencyScore"].max()

patients["UtilityScore"] = 1 - patients["EPTSScore"]/100

patients.to_csv("patients.csv")
```
